```python
# Problem 1
# Create a DataFrame:
# data = {
#  "Name": ["A", "B", "C", "D", "E, "F", "G", "H"],
#  "Marks": [45, 70, 85, 60, 58 , 78, 81, 39]
# }
#  Display students who scored more than 60.
#  Display students who scored less than or equal to 60
```

```python
import pandas as pd
data = {
    "Name": ["A","B","C","D","E","F","G","H"],
    "Marks": [45,70,85,60,58,78,81,39],
}
df = pd.DataFrame(data)
print(df)
```

```python
# Students who scored more than 60
data = df[df['Marks'] > 60]
print(data)
print()
```

```python
#Students who scored less than or equal to 60
data = df[df['Marks'] <= 60]
print(data)
```

```python
# Problem -2
# Given:
# data = {
#  "Name": ["A", "B", "C", "D", "E"],
#  "Marks": [80, None, 45, 90, 60],
#  "Attendance": [85, 70, 60, 95, None]
# }
# Tasks:
# 1. Create a DataFrame.
# 2. Replace missing values in "Marks" with the mean of that column.
# 3. Drop rows where "Attendance" is missing.
# 4. Create a new column "Result":
# o "Pass" if Marks ≥ 50
# o "Fail" otherwise
# 5. Display students who passed.

```

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [80, None, 45, 90, 60],
    "Attendance": [85, 70, 60, 95, None]
}
```

```python
# Original DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
#Replaced missing Marks with mean value
marks_mean = df['Marks'].mean()
df['Marks'] = df['Marks'].fillna(marks_mean)
print(df)
```

```python
#Dropped rows with missing Attendance
df = df[df['Attendance'].notna()]
print(df)
```

```python
#Added Result column:
import numpy as np
df['Result'] = np.where(df['Marks'] >= 50, 'Pass', 'Fail')
print(df)
```

```python
#Students who passed:
passed_students = df[df['Result'] == 'Pass']
print(passed_students)
```
