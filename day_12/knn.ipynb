```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
df=load_breast_cancer()
```

```python
df
```

```python
df1=pd.DataFrame(np.c_[df.data,df.target],columns=[list(df.feature_names)+['target']])
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
x=df1.iloc[:,0:-1].values
y=df1.iloc[:,-1].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=50,test_size=.75)
```

```python
print("shape of x_tesst",x_test.shape)
print("shape of x_train:",x_train.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
cr=KNeighborsClassifier(n_neighbors=6)
cr.fit(x_train,y_train)
```

```python
y_pred=cr.predict(x_test)
```

```python
y_pred
```

```python
cr.score(x_test,y_test)
```
