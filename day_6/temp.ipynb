```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\nk424\OneDrive\Scans\Desktop\training\Sample_Data\Temp.csv")
df.head(4)
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Days'], df['Temp'],color='g',marker='*',linestyle='--')
plt.title("Bhubaneswar City Temp")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.show()
```

```python
# BAR PLOT

plt.bar(df['Days'], df['Temp'],color='y',width=0.5)
plt.title("Bhubaneswar City Temp")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.show()
```

```python
# pie plot
x=np.array([35,25,20,15,20,43,34,23,12])
l1=["x",'y','a','b','c','d','d','d','e']
plt.pie(x, labels=l1)
plt.show()
```
